```python
import pandas as pd

# Membaca data dari CSV
data = pd.read_csv('./data-pom2.csv', sep=';', index_col=False)

# Membuat DataFrame baru dengan baris "TOTAL" yang menggabungkan 4 sel
total_row = pd.DataFrame({'Pelanggan': ['TOTAL'], 'Waktu Kedatangan': [''], 'Mulai Pelayanan': [''], 'Selesai Pelayanan': ['']})

# Menggabungkan DataFrame "TOTAL" dengan DataFrame asli
data = pd.concat([data, total_row], ignore_index=True)

# Mengonversi kolom "Mulai Pelayanan" dan "Selesai Pelayanan" ke tipe data numerik
data['Mulai Pelayanan'] = pd.to_numeric(data['Mulai Pelayanan'], errors='coerce')
data['Selesai Pelayanan'] = pd.to_numeric(data['Selesai Pelayanan'], errors='coerce')

# Membuat Series "Lama Pelayanan" dengan menghitung selisih Selesai Pelayanan - Mulai Pelayanan
data['Lama Pelayanan'] = data['Selesai Pelayanan'] - data['Mulai Pelayanan']

# Membuat Series "Pelanggan Menunggu" dengan menghitung Waktu Kedatangan - Mulai Pelayanan
data['Pelanggan Menunggu'] = data['Mulai Pelayanan'] - data['Waktu Kedatangan']

# Membuat Series "Petugas Idle" dengan menghitung perbedaan antara Mulai Pelayanan saat ini dengan Selesai Pelayanan sebelumnya
data['Petugas Idle'] = data['Mulai Pelayanan'].shift(-1) - data['Selesai Pelayanan']

# Menghitung total lama pelayanan, pelanggan menunggu, dan petugas idle dari semua baris kecuali baris terakhir
total_lama_pelayanan = data.iloc[:-1]['Lama Pelayanan'].sum()
total_pelanggan_menunggu = data.iloc[:-1]['Pelanggan Menunggu'].sum()
total_petugas_idle = data.iloc[:-1]['Petugas Idle'].sum()

# Memasukkan total lama pelayanan ke baris terakhir
data.iloc[-1, data.columns.get_loc('Lama Pelayanan')] = total_lama_pelayanan

# Memasukkan total pelanggan menunggu ke baris terakhir
data.iloc[-1, data.columns.get_loc('Pelanggan Menunggu')] = total_pelanggan_menunggu

# Memasukkan total petugas idle ke baris terakhir
data.iloc[-1, data.columns.get_loc('Petugas Idle')] = total_petugas_idle

data['Lama Pelayanan'] = data['Lama Pelayanan'].astype(int)
data['Pelanggan Menunggu'] = data['Pelanggan Menunggu'].astype(int)

# Menghilangkan desimal dari "Petugas Idle"
data['Petugas Idle'] = data['Petugas Idle'].fillna(0).astype(int)
data['Selesai Pelayanan'] = data['Selesai Pelayanan'].fillna(0).astype(int)
data['Mulai Pelayanan'] = data['Mulai Pelayanan'].fillna(0).astype(int)

# Menampilkan DataFrame yang sudah diperbarui
data = data.set_index('Pelanggan')
data
```
