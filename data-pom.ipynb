```python
import pandas as pd
import openpyxl as xl
```

```python
# DATA POM 1
data1 = pd.read_excel('./excel/dataPom1.xlsx', sheet_name='Sheet1')

# Menampilkan DataFrame yang sudah diperbarui
pom1 = data1.set_index('Pelanggan')
pom1

```

```python
# DATA POM 2
data2 = pd.read_excel('./excel/dataPom2.xlsx', sheet_name='Sheet1')

# Menampilkan DataFrame yang sudah diperbarui
pom2 = data2.set_index('Pelanggan')
pom2
```

```python
#TOTAL PELANGGAN
# Menghitung total pelanggan dari kedua DataFrame
total_pelanggan_pom1 = len(pom1) - 1
total_pelanggan_pom2 = len(pom2) - 1
total_pelanggan_semua_data = total_pelanggan_pom1 + total_pelanggan_pom2

# Menampilkan total pelanggan
print(f"Total Pelanggan dari POM 1: {total_pelanggan_pom1}")
print(f"Total Pelanggan dari POM 2: {total_pelanggan_pom2}")
print(f"Total Pelanggan dari Semua Data POM: {total_pelanggan_semua_data}\n")


#TOTAL PELAYANAN
# Menghitung total pelayanan dari kedua DataFrame
total_pelayanan_pom1 = pom1.iloc[-1]['Lama Pelayanan']
total_pelayanan_pom2 = pom2.iloc[-1]['Lama Pelayanan']
total_pelayanan_semua_data = total_pelayanan_pom1 + total_pelayanan_pom2

# Menampilkan total pelayanan
print(f"Total Pelayanan dari POM 1: {total_pelayanan_pom1.astype(int)} menit")
print(f"Total Pelayanan dari POM 2: {total_pelayanan_pom2.astype(int)} menit")
print(f"Total Pelayanan dari Semua Data POM: {total_pelayanan_semua_data.astype(int)} menit\n")


#TOTAL PELANGGAN MENUNGGU
# Menghitung total pelayanan dari kedua DataFrame
total_pelanggan_menunggu_pom1 = pom1.iloc[-1]['Pelanggan Menunggu']
total_pelanggan_menunggu_pom2 = pom2.iloc[-1]['Pelanggan Menunggu']
total_pelanggan_menunggu_semua_pom = total_pelanggan_menunggu_pom1 + total_pelanggan_menunggu_pom2

# Menampilkan total pelayanan
print(f"Total Pelanggan Menunggu dari POM 1: {total_pelanggan_menunggu_pom1.astype(int)} menit")
print(f"Total Pelanggan Menunggu dari POM 2: {total_pelanggan_menunggu_pom2.astype(int)} menit")
print(f"Total Pelanggan Menunggu dari Semua Data POM: {total_pelanggan_menunggu_semua_pom.astype(int)} menit\n")


#TOTAL PETUGAS IDLE
# Mengambil nilai total petugas idle dari baris terakhir DataFrame
total_petugas_idle_terakhir_pom1 = pom1.iloc[-1]['Petugas Idle']
total_petugas_idle_terakhir_pom2 = pom2.iloc[-1]['Petugas Idle']
total_petugas_idle = total_petugas_idle_terakhir_pom1 + total_petugas_idle_terakhir_pom2

# Menampilkan nilai total petugas idle dari baris terakhir
print(f"Total Petugas Idle dari Baris Terakhir POM 1: {total_petugas_idle_terakhir_pom1.astype(int)} menit")
print(f"Total Petugas Idle dari Baris Terakhir POM 2: {total_petugas_idle_terakhir_pom2.astype(int)} menit")
print(f"Total Petugas Idle dari Semua Data POM: {total_petugas_idle.astype(int)} menit\n")


#WAKTU ANTAR KEDATANGAN
total_waktu_kedatangan = ((30*2)*60) / total_pelanggan_semua_data
print(f"Waktu antar kedatangan: {int(total_waktu_kedatangan)} detik / orang")
print(f"Waktu antar kedatangan: {int(3600/total_waktu_kedatangan)} orang / jam\n")

#WAKTU PELAYANAN
total_waktu_pelayanan = (total_pelayanan_semua_data*60) / total_pelanggan_semua_data
print(f"Waktu pelayanan: {int(total_waktu_pelayanan)} detik / orang")
print(f"Waktu pelayanan: {int(3600/total_waktu_pelayanan)} orang / jam\n")
```
